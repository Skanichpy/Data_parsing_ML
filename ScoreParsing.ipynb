```python
import requests 
from bs4 import BeautifulSoup 
import numpy as np 
import pandas as pd 
from tqdm import tqdm
from time import sleep
```

```python
path = "https://myanimelist.net/topanime.php?limit="
npages = 100
marks = []
for limit in tqdm(range(0, npages*50 + 1, 50)):
    urls = path + str(limit)
    html = requests.get(urls).text
    bs = BeautifulSoup(html)
    for i in list(reversed([el for el in range(10)])):
        RES = bs.find_all("span", class_ = "text on score-label score-" + str(i))
        marks.append(RES)    
#     marks.append([el.text for el in RES])
    sleep(1.0)
```

```python
marks
```

```python
from functools import reduce 
Rate = list(reduce(lambda x,y: x+y, marks))
AnimeRate = [float(el.text) for el in Rate if not el.text.isspace()]
AnimeRate
```

```python
Score = pd.DataFrame({
    "Score:" : AnimeRate
})
```

```python
Score.to_csv("Score.csv")
```
